```python
print(df_sales_data[['Month', 'ProductA']])
```

```python
df_market_A = (df_sales_data[['Month', 'ProductA']])
df_market_A.set_index('Month', inplace=True)
print(df_market_A)
```

```python
#Bar Chart
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
x = df_sales_data['Month']
y = df_sales_data['ProductA']
plt.barh(x,width=y)
plt.xlabel('Month')
plt.ylabel('Product Amount')
plt.title('Total product A sell (pcs) per month in 2024')
```

```python
#Line chart
x = df_sales_data['Month']
y = df_sales_data['ProductA']
plt.plot(x,y)
plt.xlabel('Month')
plt.ylabel('Product Amount')
plt.xticks(rotation=90)
plt.title('Total product A sell (pcs) per month in 2024')
```

```python
#Heat map data preparations
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#plot heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(df_market_A, annot=True, cmap="YlGnBu", cbar_kws={'label': 'Sales/Quantity'}, linewidths=.5)
plt.title('Heatmap of Product Sales by Month in 2023')
plt.xlabel('Kind of Product')
plt.ylabel('Month')
plt.xticks(rotation=45) 
plt.tight_layout()
```
